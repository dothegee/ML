## Gradient Descent

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
#SGD : Stochastic Gradient descent
```

```python
dataset = pd.read_csv('LinearRegressionData.csv')
dataset
#독립변수
X = dataset.iloc[:,:-1].values # row는 전부, column은 마지막꺼 하나빼고 다
#종속변수
y = dataset.iloc[:,-1].values # row는 전부, column은 마지막꺼 하나만

from sklearn.model_selection import train_test_split
# 4개 값 출력
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
#testset를 20퍼센트, random_state -> seed 
```

```python
sgdr = SGDRegressor()
sgdr.fit(X_train, y_train)
```

```python
plt.scatter(X_train,y_train,color = "blue") #산점도
plt.plot(X_train, sgdr.predict(X_train), color = 'green') #선 그래프
plt.title("Score by hours (train data)") # 제목
plt.xlabel('hours')# X축 이름
plt.ylabel('score')# y축 이름
plt.show()
```

```python
sgdr.coef_
```

```python
sgdr.intercept_
```

```python
sgdr.score(X_test,y_test)
```

```python
sgdr.score(X_train,y_train)
```

### SGDRegressor parameter 설정

```python
sgdr2 = SGDRegressor(max_iter=1000, eta0=1e-4, random_state=0, verbose=1)
# 지수 표기법
# 1e-3 = 0.001
# 1e+3 = 1000
#max_iter : epoch, eta0 : 학습율 (디폴트값 : 0.01), verbose : 훈련세트를 반복하면서 손실이 어떻게 줄어드는지 확인
sgdr2.fit(X_train,y_train)
```

## 데이터 시각화

```python
plt.scatter(X_train,y_train,color = "blue") #산점도
plt.plot(X_train, sgdr2.predict(X_train), color = 'green') #선 그래프
plt.title("Score by hours (train data)") # 제목
plt.xlabel('hours')# X축 이름
plt.ylabel('score')# y축 이름
plt.show()
```
