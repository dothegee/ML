## Gradient Descent

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
#SGD : Stochastic Gradient descent
```

```python
dataset = pd.read_csv('LinearRegressionData.csv')
dataset
#
X = dataset.iloc[:,:-1].values # row , column   
#
y = dataset.iloc[:,-1].values # row , column  

from sklearn.model_selection import train_test_split
# 4  
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
#testset 20, random_state -> seed 
```

```python
sgdr = SGDRegressor()
sgdr.fit(X_train, y_train)
```

```python
plt.scatter(X_train,y_train,color = "blue") #
plt.plot(X_train, sgdr.predict(X_train), color = 'green') # 
plt.title("Score by hours (train data)") # 
plt.xlabel('hours')# X 
plt.ylabel('score')# y 
plt.show()
```

```python
sgdr.coef_
```

```python
sgdr.intercept_
```

```python
sgdr.score(X_test,y_test)
```

```python
sgdr.score(X_train,y_train)
```

### SGDRegressor parameter 

```python
sgdr2 = SGDRegressor(max_iter=1000, eta0=1e-4, random_state=0, verbose=1)
#  
# 1e-3 = 0.001
# 1e+3 = 1000
#max_iter : epoch, eta0 :  ( : 0.01), verbose :      
sgdr2.fit(X_train,y_train)
```

```python
plt.scatter(X_train,y_train,color = "blue") #
plt.plot(X_train, sgdr2.predict(X_train), color = 'green') # 
plt.title("Score by hours (train data)") # 
plt.xlabel('hours')# X 
plt.ylabel('score')# y 
plt.show()
```
